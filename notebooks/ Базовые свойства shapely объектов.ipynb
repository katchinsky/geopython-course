```python
from pyproj import CRS
```

```python
crs1 = CRS.from_epsg(4326)
crs2 = CRS.from_string("epsg:4326")
crs3 = CRS.from_proj4("+proj=latlon")
crs4 = CRS.from_user_input(("epsg", "4326"))
crs5 = CRS.from_string("4326") # отсутствует на видео, так как появился в последних версиях библиотеки
```

```python
print("crs1:{}\ncrs2:{}\ncrs3:{}\ncrs4:{}\ncrs5:{}".format(crs1, crs2, crs3, crs4, crs5))
```

###  <span style='font-family:Georgia;background-color:#00ffff;'> Создание объекта трансформер для смены проекции. Общие шаги для точки и полигона </span>
**1. Определяем EPSG код для UTM-проекции по координатам**

```python
import math
```

```python
def convert_wgs_to_utm(lon: float, lat: float):
    """Based on lat and lng, return best utm epsg-code"""
    utm_band = str((math.floor((lon + 180) / 6) % 60) + 1)
    if len(utm_band) == 1:
        utm_band = "0" + utm_band
    if lat >= 0:
        epsg_code = "326" + utm_band
        return epsg_code
    epsg_code = "327" + utm_band
    return epsg_code
```

```python
convert_wgs_to_utm(lon=pin_tour_eiffel.y, lat=pin_tour_eiffel.x)
```

```python
crs_paris_str = "epsg:" + convert_wgs_to_utm(
    lon=pin_tour_eiffel.y, lat=pin_tour_eiffel.x
)
crs_paris = CRS.from_user_input(crs_paris_str)
```

```python
crs_paris
```

**2. Создаем объект трансформер**

```python
crs1
```

```python
crs_paris
```

```python
from pyproj import Transformer

transformer = Transformer.from_crs(crs1, crs_paris)
```

##  <span style='font-family:Georgia;background-color:yellow;'> Буфер </span>

<span style="font-family:Georgia; font-size:16px"> Буфер — это геометрия, которая охватывает все точки на заданном расстоянии от входной геометрии. В shapely буфер вычисляется с использованием метода .buffer с заданным расстоянием. </span>

<img src="https://github.com/Ines2607/geoanalytics_python_rus/blob/main/buffer_exp.png?raw=true"
     alt="Markdown Monster icon"
     style="float: left; margin-right: 10px;width:570px;height:250px;" />

```python
from matplotlib import pyplot as plt
from shapely.geometry import LineString
```

```python
line = LineString([(0, 0), (1, 1), (0, 2), (2, 2), (3, 1), (1, 0)])
```

<span style="font-family:Georgia; font-weight:bold; font-size:16px"> Левосторонний и правосторонний буферы </span>

```python
import matplotlib
```

```python


fig, ax = plt.subplots(1,1, figsize=(3,3))



right_hand_side = line.buffer(-0.5, single_sided=True)
patch1 = matplotlib.patches.Polygon(right_hand_side.exterior.coords, closed=True,)

ax.add_patch(patch1)

ax.plot(*line.xy, color='red')


plt.show()

```

<span style="font-family:Georgia; font-weight:bold; font-size:16px"> Cap style и Join Style </span> 

*Из документации:*
 - **cap_style** : int, optional      
    The styles of caps are:    
    CAP_STYLE.round (1),    
    CAP_STYLE.flat(2),    
    and CAP_STYLE.square (3).   
    
 - **join_style** : int, optional    
    The styles of joins between offset segments are:   
    JOIN_STYLE.round (1),    
    JOIN_STYLE.mitre (2), and   
    JOIN_STYLE.bevel (3).


<span style="font-family:Georgia;"> Документация метода буфер : https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoSeries.buffer.html , http://shapely.readthedocs.io/en/latest/manual.html#object.buffer </span>

```python
fig, ax = plt.subplots(3, 3, figsize=(18, 18))

for join_st in range(1, 4):
    for cap_st in range(1, 4):
        left_hand_side = line.buffer(0.2, cap_style=cap_st, join_style=join_st)

        patch1 = matplotlib.patches.Polygon(left_hand_side.exterior.coords, alpha=0.5)

        ax[join_st - 1, cap_st - 1].add_patch(patch1)
        ax[join_st - 1, cap_st - 1].plot(*line.xy, color="red")
        ax[join_st - 1, cap_st - 1].set_title(
            "Cap style:{}, join style:{}".format(cap_st, join_st)
        )
plt.show()
```
