```python
# функция из предыдущего ноутбук
import math
def convert_wgs_to_utm(lon: float, lat: float):
    """Based on lat and lng, return best utm epsg-code"""
    utm_band = str((math.floor((lon + 180) / 6) % 60) + 1)
    if len(utm_band) == 1:
        utm_band = "0" + utm_band
    if lat >= 0:
        epsg_code = "326" + utm_band
        return epsg_code
    epsg_code = "327" + utm_band
    return epsg_code
```

```python
crs_str = "epsg:" + convert_wgs_to_utm(lat=pin_point.x, lon=pin_point.y)
print(crs_str)
crs_fin = CRS.from_user_input(crs_str)
crs_fin
```

 **<span style='font-family:Georgia'> 5. Проецируем точку в UTM  </span>**

```python
from shapely.geometry import Point
```

```python
my_loc = Point([55.78956365881177, 37.606846598229254])
```

```python
# Создаем функцию из ранее выполненных шагов, чтобы проецировать точку в нужную систему координат
def receive_new_coords_point(point):
    init_crs = "epsg:4326"
    init_crs = CRS.from_string(init_crs)

    crs_str = "epsg:" + convert_wgs_to_utm(lon=point.y, lat=point.x)
    crs_fin = CRS.from_user_input(crs_str)
    print(crs_str)

    transformer = Transformer.from_crs(init_crs, crs_fin)
    new_coords = transformer.transform(xx=point.x, yy=point.y)

    return Point(new_coords)
```

```python
my_loc_utm = receive_new_coords_point(my_loc)
```

```python
my_loc_utm.coords.xy
```

```python
gdf_metro_utm['min_dist'] = -1.0
for index, row in tqdm(gdf_metro_utm.iterrows()):
    name = row['NameOfStation']
    geom = row['geometry']
    
    dist_array = gdf_metro_utm[gdf_metro_utm['NameOfStation'] != name]['geometry'].distance(geom)
   
    gdf_metro_utm.loc[index, 'min_dist'] = dist_array.min()
gdf_metro_utm
```

##  <span style='font-family:Georgia; background-color:yellow;'> Проверка принадлежности объектов буферу</span>

Шаги:
    
    1. Строим от каждой станции буфер 2 км
    2. Для каждой пары станций проверяем, входит ли одна в 2км радиус другой станции
    3. Считаем для каждой станции, сколько выходов со станций с другим названием входят в ее 2км буфер
    4. Выводим только те станции, где у всех выходов кол-во соседних станций = 0

<span style=" font-family:Georgia; font-weight: bold; "> 1. Строим от каждой станции буфер 2 км </span>

```python
gdf_metro_utm["geometry_buffer"] = gdf_metro_utm["geometry"].buffer(2000)
gdf_metro_utm.head()
```

<span style=" font-family:Georgia; font-weight: bold; ">  Шаги в цикле:  </span>    

<span style=" font-family:Georgia; font-weight: bold; "> 2. Для каждой пары станций проверяем, входит ли одна в 2км радиус другой станции
</span>   
<span style=" font-family:Georgia; font-weight: bold; "> 3. Считаем для каждой станции, сколько выходов со станций с другим названием входят в ее 2км буфер  
</span>

```python
gdf_metro_utm['number_neigbors_2km'] = 0

for index, row in tqdm(gdf_metro_utm.iterrows()):
    
    name = row['NameOfStation']
    geom_buffer = row['geometry_buffer']
    
    #2. Для каждой пары станций проверяем, входит ли одна в 2км радиус другой станции    
    flag_array = gdf_metro_utm[gdf_metro_utm['NameOfStation'] != name]['geometry'].within(geom_buffer) 
    
    #3. Считаем для каждой станции, сколько выходов со станций с другим названием входят в ее 2км буфер   
    gdf_metro_utm.loc[index,'number_neigbors_2km'] = sum(flag_array)
gdf_metro_utm.head()
```


<span style=" font-family:Georgia; font-weight: bold; "> 4. Выводим только те станции, где у всех выходов кол-во соседних станций = 0 </span>

Cделаем это, проверив максимальное значение соседей в 2 км для каждого названия станции. Оно должно должно быть равно 0. Если больше, это значит, что хотя бы у 1 из выходов нашелся выход из другой станции в радиусе 2 км)

```python
series_num_neigbor = gdf_metro_utm.groupby(["NameOfStation"])[
    "number_neigbors_2km"
].max()
series_num_neigbor.sort_values()
```

```python
series_num_neigbor[series_num_neigbor.values == 0]
```

```python
print(
    "Список наиболее удаленных станций: {}".format(
        series_num_neigbor[series_num_neigbor == 0].index.tolist()
    )
)
```

Шаги:

    2. Формируем 2 geodataframe формата "геометрия-имя станции- global_id": в 1-ом геометрия - точка станции, во 2-ом геометрия - полигон буфера

В видео Практическое занятие 8 указаны 3 доступные операции, однако начиная с geopandas>0.10 их кол-во увеличилось.

<span style=" font-family:Georgia; font-weight: bold; "> 1. Строим от каждой станции буфер 2 км </span>

```python
gdf_metro_utm["geometry_buffer"] = gdf_metro_utm["geometry"].buffer(2000)
```

<span style=" font-family:Georgia; font-weight: bold; ">  2. Формируем 2 geodataframes формата "геометрия-имя станции- global_id": в 1-ом геометрия - точка станции, во 2-ом геометрия - полигон буфера</span>

```python
gdf_metro_utm2 = gdf_metro_utm[["geometry", "NameOfStation", "global_id"]]
```

```python
gdf_metro_utm_buffer = gdf_metro_utm.set_geometry("geometry_buffer")[
    ["geometry_buffer", "NameOfStation", "global_id"]
]
```

<span style=" font-family:Georgia; font-weight: bold; ">    3. Сцепляем их с помощью spatial join. Логика : buffer.contains(point)
</span>

<span style=" font-family:Georgia"> Сцепляем оба датасета через функцию spatial join.
    Порядок указания датасетов важен, потому что показывает по значениям какой таблицы мы фильтруем сцепку, то есть какая таблица основная - по ней фильтр, а какая добавочная.

В отличие от pandas.join добавляется аттрибут 'predicate'( в ранних версиях 'op')

<span style=" font-family:Georgia"> **Важно**: остается только одна колонка с геометрией - из таблицы которая указана слева
</span>

```python
%%time

gdf_metro_utm_pairs2km = gdf_metro_utm_buffer.sjoin(\
                                                    gdf_metro_utm2,
                                                    how ='left',
                                                    predicate ='contains',
                                                    lsuffix ='main',
                                                    rsuffix ='neighbor'
                                                   )
```

```python
gdf_metro_utm_pairs2km.columns
```

```python
gdf_metro_utm_pairs2km
```

```python
gdf_metro_utm_pairs2km["flag_dif_station"] = (gdf_metro_utm_pairs2km["NameOfStation_main"] != gdf_metro_utm_pairs2km["NameOfStation_neighbor"])
```

```python
gdf_metro_utm_pairs2km["flag_dif_station"].value_counts()
```

```python
df_stat_enter_num = (
    gdf_metro_utm_pairs2km.groupby(["NameOfStation_main", "global_id_main"])["flag_dif_station"]
    .sum()
    .to_frame()
)
```
