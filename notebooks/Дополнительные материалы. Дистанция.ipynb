Для рачета расстояния между геометрическими объектами используют  Эвклидово, Манхэттенское, Чебышева, Гаверсин расстояния.

```python


# defining the points
point_1 = Point(1, 2)
point_2 = Point(2, 4)
point_1, point_2
```

<span style="font-family:Georgia; font-weight:bold; font-size:18px">Эвклидово расстояние</span>   

Эвклидово расстояние - метрика в евклидовом пространстве — расстояние между двумя точками евклидова пространства, вычисляемое по теореме Пифагора.

```python
# Преобразуем точки в координаты для передачи в distance.euclidean
eucledian_distance = distance.euclidean((point_1.x, point_1.y), (point_2.x, point_2.y))
print('Эвклидово расстояние', point_1, 'and', point_2, 'is: {:.3f}'.format(eucledian_distance))
```

<span style="font-family:Georgia; font-weight:bold; font-size:18px">Манхэттенское расстояние</span>

Манхэттенское расстояние, часто называемое расстоянием такси или расстоянием городского квартала, вычисляет расстояние между векторами с действительными значениями. Если представить шахматную доску, то это минимальное расстояние, которое проходит ладья между 2-мя полями.То есть это истинная дистанция объекта, который старается двигаться оптимально и которому разрешено поворачивать только под прямым углом.



Если Эвклидово расстояние - это самая оптимистичная оценка движения объекта, то манхэттенское используют как оценка сверху. Например, используют, чтобы оценить время необходимое скорой помощи, чтобы довезти пациента до больницы или пешему курьеру,чтобы доставить поссылку. Если курьер двигается по треку хуже манхэттенского значит либо он "фродит",либо с дорожной сетью что-то не то.

```python
# Преобразуем точки в координаты для передачи в distance.cityblock
manhattan_distance = distance.cityblock((point_1.x, point_1.y), (point_2.x, point_2.y))
print('Manhattan Distance b/w', point_1, 'and', point_2, 'is: ', manhattan_distance)
```

<span style="font-family:Georgia; font-weight:bold; font-size:18px">Расстояние Чебышева</span>

Расстояние Чебышева определяется как наибольшая разница между двумя векторами в любом координатном измерении. Другими словами, это просто максимальное расстояние вдоль одной оси. Из-за своей природы его часто называют расстоянием шахматной доски, поскольку минимальное количество ходов, необходимое королю для перехода с одного поля на другое, равно расстоянию Чебышева.


Расстояние Чебышева иногда используется в складской логистике, так как оно эффективно измеряет время, необходимое мостовому крану для перемещения объекта (поскольку кран может перемещаться по осям x и y одновременно, но с одинаковой скоростью по каждой из них).

```python
# Преобразуем точки в координаты для передачи в distance.chebyshev
chebyshev_distance = distance.chebyshev((point_1.x, point_1.y), (point_2.x, point_2.y))
print('Расстояние Чебышева', point_1, 'and', point_2, 'is: ', chebyshev_distance)
```

<span style="font-family:Georgia; font-weight:bold; font-size:18px"> Расстояние по гаверсинусу ( Haversine distance) </span>

Расстояние по гаверсинусу (или по большому кругу) — это угловое расстояние между двумя точками на поверхности сферы. Первая координата каждой точки принимается за широту, вторая за долготу, указанную в радианах. Размерность данных должна быть равна 2.

Используется при расчете расстояний между двумя удаленными объектами, например если вы хотите оценить дистанцию между Москвой и Владивостоком.
Подробнее: https://medium.com/ninjavan-tech/the-magic-of-haversine-distance-1c4e1641d880

```python
import math
def haversine(lat1, lon1, lat2, lon2):
     
    # distance between latitudes
    # and longitudes
    dLat = (lat2 - lat1) * math.pi / 180.0
    dLon = (lon2 - lon1) * math.pi / 180.0
 
    # convert to radians
    lat1 = (lat1) * math.pi / 180.0
    lat2 = (lat2) * math.pi / 180.0
 
    # apply formulae
    a = (pow(math.sin(dLat / 2), 2) +
         pow(math.sin(dLon / 2), 2) *
             math.cos(lat1) * math.cos(lat2));
    rad = 6371
    c = 2 * math.asin(math.sqrt(a))
    return rad * c
 
```

```python
print('Расстояние по гаверсинусу',haversine(point_1.x,point_1.y, point_2.x, point_2.y))
```
